**5. Find out how often the price from Bookbot is between the 25% cheapest and most expensive offers.**

```python
# test data
import pandas as pd

# Test data
data = {
    'ISBN13': ['9780132350884', '9780132350884', '9780132350884', 
               '9781593276034', '9781593276034', '9781593276034',
               '9780160723001', '9780160723001', '9780160723001',
              '9780160723002', '9780160723002', '9780160723002',
              ],
    'Title': ['Pride and Prejudice', 'Pride and Prejudice', 'Pride and Prejudice', 
              'Python Crash Course', 'Python Crash Course', 'Python Crash Course', 
              'Python Data Science Handbook', 'Python Data Science Handbook', 'Python Data Science Handbook',
             'On The Road', 'On The Road', 'On The Road', 
             ],
    'Provider': ['Bookbot', 'Barnes & Noble', 'Amazon', 
                 'O Reilly', 'O Reilly', 'Bookbot',
                 'Barnes & Noble', 'Amazon', 'Bookbot', 
                'Barnes & Noble', 'Amazon', 'Amazon',
                ],
    'Prices': [60, 50, 70, 140, 100, 120, 100, 120, 140, 100, 150, 200]
}
# Creating DataFrame
df = pd.DataFrame(data)
# Displaying DataFrame
print(df)
```

```python
# Filter the DataFrame to include only rows where Bookbot is mentioned at least once
bookbot_isbns = df[df['Provider'] == 'Bookbot']['ISBN13'].unique()
print(bookbot_isbns)
bookbot_df = df[df['ISBN13'].isin(bookbot_isbns)]
print(bookbot_df)

# Group by ISBN and calculate the 25th and 75th percentile for each ISBN
percentiles = bookbot_df.groupby('ISBN13')['Prices'].quantile([0.25, 0.75]).unstack()
print(percentiles)
print(type(percentiles))

# only include rows where the Provider is 'Bookbot'
bookbot_prices = df[df['Provider'] == 'Bookbot'][['ISBN13', 'Prices']]
print(bookbot_prices)

# Merge dataframes on ISBN13
merged_df = pd.merge(bookbot_prices, percentiles, on='ISBN13', suffixes=('_price', '_percentile'))
print(merged_df)

# # Compare prices with percentiles and check if price is smaller than 25th percentile or bigger than 75th percentile
merged_df['is_outlier'] = (merged_df['Prices'] < merged_df[0.25]) | (merged_df['Prices'] > merged_df[0.75])
print(merged_df)

# Count number of outliers
num_outliers = merged_df['is_outlier'].sum()

print("Number of outliers:", num_outliers)
```

**6. Identify sellers who are more likely to have a lower price than BookBot.**

**7. Calculate the average deviation of Bookbot's offers from the lowest price**

```python
# test data
import pandas as pd

# Test data
data = {
    'ISBN13': ['9780132350884', '9780132350884', '9780132350884', 
               '9781593276034', '9781593276034', '9781593276034',
               '9780160723001', '9780160723001', '9780160723001',
              '9780160723002', '9780160723002', '9780160723002',
              ],
    'Title': ['Pride and Prejudice', 'Pride and Prejudice', 'Pride and Prejudice', 
              'Python Crash Course', 'Python Crash Course', 'Python Crash Course', 
              'Python Data Science Handbook', 'Python Data Science Handbook', 'Python Data Science Handbook',
             'On The Road', 'On The Road', 'On The Road', 
             ],
    'Provider': ['Bookbot', 'Barnes & Noble', 'Amazon', 
                 'O Reilly', 'O Reilly', 'Bookbot',
                 'Barnes & Noble', 'Amazon', 'Bookbot', 
                'Barnes & Noble', 'Amazon', 'Amazon',
                ],
    'Prices': [60, 50, 70, 140, 100, 120, 100, 120, 140, 100, 150, 200]
}
# Creating DataFrame
df = pd.DataFrame(data)
# Displaying DataFrame
print(df)
```

```python
# Get ISBNs where Bookbot Provider is present
bookbot_isbns = df[df['Provider'] == 'Bookbot']['ISBN13'].unique()
print("bookbot_isbns", bookbot_isbns)

bookbot_prices = df[df['Provider'] == 'Bookbot']['Prices'].tolist()
print("bookbot_prices", bookbot_prices)

# Lowest price for each ISBN where Bookbot is present
lowest_prices = []
for isbn in bookbot_isbns:
    lowest_prices.append(df[df['ISBN13'] == isbn]['Prices'].min())
print("lowest_prices", lowest_prices)

# Deviation of each Bookbot's offer from the lowest price
deviations = []
for price, lowest_price in zip(bookbot_prices, lowest_prices):
    deviations.append(price - lowest_price)
print("deviations", deviations)

# Average deviation of Bookbot's offers from the lowest price
average_deviation = sum(deviations) / len(deviations)

print(f"Average deviation of Bookbot's offers from the lowest price: {average_deviation}")
```

**8. Analyse the relative representation of foreign offers in the downloaded data.**

```python
# test data
import pandas as pd

# Test data
data = {
    'ISBN13': ['9780132350884', '9780141439563', '9780061120084', '9781593276034', '9781449369415'],
    'Title': ['Clean Code: A Handbook of Agile Software Craftsmanship', 
              'Pride and Prejudice', 
              'The Pragmatic Programmer: Your Journey to Mastery', 
              'Python Crash Course', 
              'Python Data Science Handbook'],
    'Provider': ['Amazon', 'Barnes & Noble', 'Amazon', 'O Reilly', 'O Reilly'],
    'Country': ['USA', 'Germany', 'Japan', 'Czech Republic', 'USA'],
    'Prices': [29.99, 7.99, 30.49, 22.99, 40.0]
}
# Creating DataFrame
df = pd.DataFrame(data)
# Displaying DataFrame
print(df)
```

```python
# Extracting foreign offers (excluding Czech Republic)
foreign_offers = df[df['Country'] != 'Czech Republic']

# Counting the number of foreign offers
foreign_offers_count = len(foreign_offers)

# Total number of offers
total_offers = len(df)

# Calculating the relative representation of foreign offers
relative_representation = foreign_offers_count / total_offers * 100

# Displaying the result
print("\nRelative representation of foreign offers:", relative_representation, "%")
```
